```python
from bs4 import BeautifulSoup

import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_bicycle-sharing_systems'

page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find('table', class_ ='wikitable sortable sticky-header')
```

```python
table = soup.find('table')
```

```python
titles = table.find_all('th')
```

```python
titles
```

```python
table_headers = [headers.text.strip() for headers in titles]
table_headers
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_headers)
```

```python
df
```
